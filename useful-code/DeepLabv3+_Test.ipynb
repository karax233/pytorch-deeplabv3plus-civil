# **Merge images and masks**

```python
import shutil
import os
path = '/content/drive/MyDrive/Lab2020/Liu/Deeplab_pytorch/Datasets/Katrina03' # dataset path

image_path = os.path.join(path, 'image')
mask_path = os.path.join(path, 'predicted')
output_path =  os.path.join(path, 'merged')

images = os.listdir(image_path)
masks = os.listdir(mask_path)


new_image = []
for x in images:
  if x[-8:-4] != 'mask':
    new_image.append(x)
images = new_image

print(len(images), len(masks))
assert len(images) == len(masks)
```

```python
# merge images with predicted masks 
if os.path.exists(output_path):
    shutil.rmtree(output_path)
if not os.path.exists(output_path):
    os.makedirs(output_path)

from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
import cv2

#image1 is image
#image2 is mask
n = 0
sum = len(images)

for (image, mask) in zip(images, masks):
  image1 = Image.open(os.path.join(image_path, image))
  image2 = Image.open(os.path.join(mask_path, mask))
  image1 = image1.convert('RGBA')
  image2 = image2.convert('RGBA')

#blended_img = img1 * (1 – alpha) + img2* alpha
  img = Image.blend(image1,image2,0.5)
  bright_enhancer = ImageEnhance.Brightness(img)

# parameters = 2
  bright_img = bright_enhancer.enhance(2)
  bright_img.save(os.path.join(output_path, image[:-4] + '.png'))
  n += 1
  print("Number: {0} / {1}".format(n, sum))
```
